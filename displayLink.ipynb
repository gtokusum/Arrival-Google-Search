```python
import pandas as pd
import json
import requests
from dotenv import load_dotenv
import os
```

```python
API_KEY = os.environ['GOOGLE_API_KEY']
SEARCH_ENGINE_ID = os.environ['SEARCH_ENGINE_ID']
```

```python
df = pd.read_excel("websites.xlsx")
df.head()
```

```python
for i in range(len(df)):
    print(df.loc[i]['Forbes.com'])
```

```python
def search(df,api_key,search_engine_id):
    data = []
    for i in range(len(df)):
        name = df.loc[i]['Forbes.com']
        search_url = f'https://www.googleapis.com/customsearch/v1?q={name}&cx={search_engine_id}&key={api_key}'
        response = requests.get(search_url)
        data.append(json.loads(response.text))
    return data
```

```python
datajson = search(df,API_KEY,SEARCH_ENGINE_ID)
```
